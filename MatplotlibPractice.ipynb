# **IMPORTANT KEY TERMS**
Data Point : (x,y)                                                                
X-axis / Y-axis : input / output                                                 
Figure : entire canas where plots lie                                            
Axes : Graph Box / drawin inside a figure                                        
Plot : Actual line, graph, bar drawn inside axes                                 
Marker : symbol used to define data points, O = circle, S = sqaure, D = diamond, X = cross, + = plus                                                     
Line : continuos connection between data points                                  
Line Style : - solid line, -- dashed line, : dotted line, -. dash-dot line       
Color : defines colour of all elements                                           
Legend : box defining which line and colour represent what value                   
Label : tells the purpose of X-axis and Y-axis                                   
Title : Gives title to your plot                                                 
Grid : grid lines enhance readability and data visualization                      
Function : block of code performing some task

```python
months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months, sales, color='blue', linestyle='--', linewidth = 2, marker = 'o', label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Report')
plt.legend(loc = 'upper left', fontsize=12)
plt.grid(color = 'grey', linestyle = ':', linewidth=1)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

```python
import matplotlib.pyplot as plt
#vertical bar graph using bar()
#plt.bar(x, y, color='color_name, width = value, label='label_name')
product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product, sales, color='DarkSeaGreen', label='2025 sales data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend(loc = 'upper left', fontsize=8)
plt.show()
```

```python
#Horizontal bar graph using : barh()
#plt.bar(x, y, color='color_name, width = value, label='label_name')
product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.barh(product, sales, color='RosyBrown', label='2025 sales data')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend(loc = 'upper left', fontsize=8)
plt.show()
```

```python
#Pie Chart
#plt.pie(values, labels=[label_list], color=[color_list], autopct)
regions = ["north", "south","east","west"]
revenue = [3000,2000,1500,1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['Salmon','Olive','Navy','SlateGrey'])
plt.xlabel('Regions')
plt.ylabel('Revenue')
plt.title('Region Revenue')
plt.show()
```

```python
#Histogram
#plt.hist(data, bins=numbers_of_bins, color='colorname', egdecolor='colorname')
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color='lavender', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.title('Test Score Distribution')
plt.show()
```

```python
#Scatter Plots
#plt.scatter(x, y, color='colorname, marker='markerstyle, label='lablename')
hours_studied=[1,2,3,4,5,6,7,8]
exam_Scores=[50,55,60,65,70,75,80,85]

plt.scatter(hours_studied, exam_Scores, color='MidnightBlue', marker='^', label="Student Data")
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between Study Hours & Exam Score')
plt.legend(loc = 'upper left', fontsize=8)
plt.grid(True)
plt.show()
```

# SubPlot
Multiple plots within single canvas.

```python
#plt.subplot(nrows, ncols, index)
x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) #1 row, 2 column, 1st subplot
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2) #1 row, 2 column, 2nd subplot
plt.bar(x,y)
plt.title("Bar Chart")

plt.show()
```

```python
#fig, ax=plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax=plt.subplots(1,2, figsize=(10,5))

x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title("Line Chart")

ax[1].bar(x,y)
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison of two charts")
plt.tight_layout()
plt.show()
```
